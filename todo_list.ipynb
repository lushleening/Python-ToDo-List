#### todolist_menu function
The function todolist_menu is to display the To-Do list menu to the users for them to choose an option when the programme is executed.

```python
def todolist_menu():
    print("\n*****To-Do List*****\n")
    print("1. Add Task")
    print("2. View All Tasks")
    print("3. Delete Task")
    print("4. Mark Task as Done")
    print("5. View Pending Tasks")
    print("6. Update Tasks")
    print("7. Exit")
```

#### AddTask function
- In AddTask(tasks) function, it takes a parameter called tasks which is a list used to store all user input regarding tasks and description as well as its status.
- User is required to enter a task along with a description. The output will then be appended to the list.
- A statement stating that the task has been added will be printed once task and description has been entered.

```python
def AddTask(tasks):
    while True:
        task = input("Enter a task: ").strip()
        if not task:
            print("Task cannot be empty. Please re-enter a task\n")
            continue
        description = input("Enter a description: ")
        tasks.append({"task":task, "description":description, "completed": False})
        print(f"\nTask '{task}' has been added successfully!")
        return
```

#### ViewTasks function
- In this function, a list of all the tasks and description will be printed along with its status which is either done or pending using the enumerate function. 
- If there are no tasks, a statement stating that the list is empty will be printed.

```python
def ViewTasks(tasks):
    if tasks:
        print("\n-----List of all tasks-----")
        for i, task in enumerate(tasks, start=1):
            des = task["description"]
            status = "Done" if task["completed"] else "Pending"
            print(f"\n{i}. Task: {task['task']}\n   Description: {des} \n   Status: {status}")
    else:
        print("\nYour to-do list is empty")
        return
```

#### DeleteTask function
- In this function, users are prompted to choose a task to delete.
- It uses an if-else block to check if the number that the user entered is within the range of the list. If it is not, user is prompted to re-enter a valid number.
- If users enter anything else apart from the range of numbers that is specified (for example words), the programme will state that it is not a number and keep on prompting users to re-enter.

```python
def DeleteTask(tasks):
    ViewTasks(tasks)

    while True:
        if len(tasks)!=0: 
            try:
                index = int(input(f"\nChoose a task to delete (1-{len(tasks)}): "))-1
                if 0<= index < len(tasks):
                    del tasks[index]
                    print("\nTask deleted successfully")
                    return index
                else:
                    print(f"\nChoose a valid number (1-{len(tasks)}). Please enter again")
            except ValueError:
                print("\nThis is not a number. Please enter again")
        else:
            return
```

#### MarkDone function
- In this function, users are prompted to choose a task to be marked as done.
- If users enter anything outside of the range of number specified, the programme will continuously prompt users to re-enter. (implemented using while loop)
- Then, if users try to mark a task as done after already marking it once, the programme will state to users that the task has already been marked done before.
- If all the tasks have been marked done, a statement stating that there is no pending tasks will be printed when user choose this option. It will bring the user back to the main function
- If there are no tasks, it will be stated that there are no tasks to be marked

```python
def MarkDone(tasks):
    allcompleted = all(task["completed"] for task in tasks)

    if not tasks:
        print("\nThere are no tasks to be marked")
        return
    elif allcompleted:
        print("\nAll tasks have been marked as done. There is no more tasks to be marked")
        return   
    
    ViewTasks(tasks)

    while True:
        try:
            while True:
                index = int(input(f"\nChoose a task to be marked as done (1-{len(tasks)}): "))-1
                if 0<= index < len(tasks):
                        if tasks[index]["completed"] == False:
                            tasks[index]["completed"] = True  
                            print("\nTask completed successfully!")
                            return index
                        else:
                            print("Task has already been marked as completed. Please choose another task")
                else:
                    print(f"\nChoose a valid number (1-{len(tasks)}). Please enter again")
        except ValueError:
            print("\nThis is not a number. Please enter again")
```

#### PendingTasks function
- In this function, all the pending tasks are displayed (only tasks and description will be displayed)
- If all tasks have been completed, it will be stated that there is no pending tasks
- If there isn't any tasks created, it will be stated that there is no tasks

```python
def PendingTasks(tasks):
    if tasks:
        i=1
        print("\n-----List of Pending tasks-----\n")
        for task in tasks:
            des = task["description"]
            if task["completed"] == False:
                print(f"{i}. Task: {task['task']}\n   Description: {des}\n")
                i+=1
        if i == 1:
            print("There is no pending task")

    else:
        print("\nThere are no tasks")
```

#### UpdateTasks function
- In this function UpdateTasks(tasks), users are prompted to choose which task to update based on the tasks number
- The function first checks if the number that the user enters is within the range of the list
- If it is within the range of the list, user is prompted to enter an updated task (a task that they want to modify). If it is left empty, the programme will continuously prompt users to enter a task using a while loop
- Once user enters a task, the programme will prompt user to enter a description (description can be empty as users are given the flexibility to choose whether to enter a description or not)
- The programme will then update the task and description, overwriting the original task and description
- However, if the number that the user enters is not within the range of the list, they are continuously prompted to enter a valid number
- If the user enters anything other than numbers, the programme will state that it is not a number and prompt users to enter again

```python
def UpdateTasks(tasks):
    ViewTasks(tasks)

    if not tasks:
        print("\nThere are no tasks to be updated")
        return

    while True:
        try:
            index = int(input(f"\nChoose a task to update (1-{len(tasks)}): "))-1
            if 0<= index < len(tasks):
                    while True:
                        updateTask = input("Enter your updated task: ").strip()
                        if not updateTask:
                            print("Task cannot be empty. Please re-enter a task\n")
                        else:
                            updateDescription = input("Enter your updated description: ")
                            tasks[index]['task'] = updateTask
                            tasks[index]['description'] = updateDescription
                            print("\nTask has been updated successfully")
                            return tasks, index, updateTask, updateDescription
            else:
                print(f"\nChoose a valid number (1-{len(tasks)}). Please enter again")
        except ValueError:
            print("\nThis is not a number. Please enter again")
```

#### Main function
- In the main function, the programme displays the todolist_menu which lists the menu/option for users to choose
- The programme uses an if-elif-else block that prompt users to choose an option
- The programme will execute a specific function based on the users choice
- if users choice is not between the range of numbers specified(1-7), the programme will continuously prompt the user to enter a valid input
- If user chooses to exit (choice 7), the programme will terminate

```python
def main():
    tasks = []

    while True:
        todolist_menu()
        choice = input("\nEnter your choice (1-7): ")

        if choice == '1':
            AddTask(tasks)
        elif choice == '2':
            ViewTasks(tasks)
        elif choice == '3':
            DeleteTask(tasks)
        elif choice == '4':
            MarkDone(tasks)
        elif choice == '5':
            PendingTasks(tasks)
        elif choice == '6':
            UpdateTasks(tasks)
        elif choice == '7':
            print("You have successfully exited!")
            break
        else:
            print("\nInvalid choice.Please enter between (1-7)")
 


if __name__ == "__main__":
    main()
```

# Edge Cases
1. In the AddTask function, when the user tries to enter an empty task, they will continuously be prompted to enter a task. This edge case is to ensure that there is no empty tasks.
   Instead of allowing users to continue with an empty task, the programme will state "Task cannot be empty. Please re-enter a task".
2. When a user enters a number that is not within the range specified, the programme will continuously prompt the user to enter a valid number instead of crashing.
3. When a user enters anything other than a number, the programme will state that it is not a number and prompt the user to enter a valid number.
4. If users try to view tasks from an empty list, the programme will state that there are no tasks instead of displaying norhing or throwing an error.
5. In the MarkDone function, if user has already marked a task as done before and they mark it again, the programme will state that the task has already been marked before and prompt the user to mark a task that is still pending.
6. In the MarkDone function, if all the tasks have already been marked as done by the user, a statement stating that all tasks have been marked will be displayed instead of listing all the tasks and allowing users to mark them as done again.
7. In the PendingTasks function, if all the tasks have been marked as done, the programme will state that there are no pending tasks left instead of displaying nothing or throwing an error.
